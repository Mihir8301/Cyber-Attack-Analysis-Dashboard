# Cleaning/Pre-processing the data

```python
df.describe()
```

# i) Renaming the columns name as index 0

```python
df.columns
```

```python
df.columns = df.iloc[0]
```

```python
# Drop the first row now that it's the header
df = df[1:]

# Reset the index
df = df.reset_index(drop=True)

df.head(4)
```

# ii) Checking null values and removing or manipulating it (if required)

```python
df.isna().sum()
```

<h4> We found out that columns industry, event_subtype, description, source_url, and country having null values

1. For the "industry" field, since "industry" and "industry_code" are interlinked, we will first check if there are any records with the same "industry_code" but with a non-null "industry name" in other records. If such a record exists, we will use that "industry name" to fill the null value. If no such record is found, we will replace the null value with "not available."

2. For "event_subtype," "description," and "source_url," we will replace null values with "not available" since these columns are not considered essential for analysis according to our requirements. These columns may also be dropped in the future.

3. For the "country" field, which is essential for analysis, replacing a null value with "not available" could introduce filler records. Given that there is only one null value, it is preferable to drop that record entirely.

```python
df[df['industry'].isnull()]
```

```python
# Filter out only the desired columns
filtered_df = df.loc[:,['industry_code', 'industry']]

# Create a boolean mask for duplicates, keeping the first occurrence
duplicates_mask = filtered_df.duplicated(keep='first')

# Filter out unique rows based on the mask
unique_df = filtered_df[~duplicates_mask]

unique_df.sort_values(by='industry_code')
```

```python
df.fillna({

    'industry':'Not Available',
    'event_subtype' :'Not Available',
    'description': 'Not Available',
    'source_url': 'Not Available'
},inplace=True)

df.dropna(subset='country',inplace=True)
```

```python
df.isnull().sum()
```

# iii) Checking the duplicate values and removing it (if required)

```python
df.duplicated().sum()
```

<h4> We found out that there are no duplicates

# iv) Checking the data types and making it correctly for accuracy

```python
df.info()
```

<h4> We need to change the data type for event_date to date type and industry_code to integer

```python
# Convert 'event_date' column to datetime
df['event_date'] = pd.to_datetime(df['event_date'])

# Convert 'industry_code' column to integer
df['industry_code'] = pd.to_numeric(df['industry_code'], downcast='integer')
```

```python
df.dtypes
```

# v) Maniplulating columns based on the requirements

```python
df.columns
```

<h4> We need to rearrange columns for better visiblity of the data and need to add some new columns to fullfill the requirements

```python
df['event_date']
```

```python
# Extract the year
df['event_year'] = df['event_date'].dt.year

df
```

```python
# New column order
new_order =['slug', 'event_date','event_year','country', 'organization', 'industry', 'industry_code','event_type', 'event_subtype', 'motive', 
       'description', 'actor', 'actor_type','actor_country','source_url']


# Rearrange columns
df = df[new_order]
```

```python
#Renaming certain columns

df.rename(columns={
    'country': 'affected_country',
    'organization': 'affected_organization',
    'industry': 'affected_industry',
    'industry_code' : 'afftected_industry_code'
}, inplace=True)

df
```

# vi) Exporting cleaned dataset to excel

```python
# Export DataFrame to an Excel file
df.to_excel('Output_Cleaned_cyber_attack_dataset02.xlsx', index=False)
```
